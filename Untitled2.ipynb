```python
# 1. PERSIAPAN LINGKUNGAN DAN PEMUATAN DATA
# ------------------------------------------------------------------------------

# Import Pustaka (Libraries) yang Dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files # Pustaka untuk unggah file interaktif

# Atur gaya visualisasi default
sns.set_style("whitegrid")

nama_file = 'titanic_clean (2).xlsx'

# --- Unggah File dan Pemuatan Data ---
print("--- Mulai Proses Unggah dan Pemuatan Data ---")
print(f"Jendela pop-up akan muncul. Silakan unggah file: [{nama_file}]")

try:
    # Unggah Interaktif
    uploaded = files.upload()
    print("✅ Unggahan file ke sesi Colab berhasil diselesaikan.")

    # Pemuatan Data (Menggunakan read_excel karena format file asli adalah XLSX)
    df = pd.read_excel(nama_file)
    print("="*60)
    print("✅ Data berhasil dimuat ke dalam DataFrame 'df'.")
    print("="*60)
    print("Lima baris pertama:")
    print(df.head())
    print("\nInformasi Data (Dtype dan Missing Values):")
    df.info()

except FileNotFoundError:
    print(f"❌ ERROR: File '{nama_file}' tidak ditemukan.")
    raise SystemExit("Pemuatan data gagal. Harap pastikan file sudah diunggah dan namanya benar.")
except Exception as e:
    print(f"❌ ERROR: Terjadi kesalahan saat membaca file. Detail: {e}")
    raise SystemExit("Pemuatan data gagal.")


# ------------------------------------------------------------------------------
# 2. OBJEKTIF A: ANALISIS STATISTIK DESKRIPTIF
# ------------------------------------------------------------------------------

kolom_numerik = ['Survived', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch']

print("\n\n" + "="*60)
print("HASIL ANALISIS STATISTIK DESKRIPTIF KUNCI")
print("="*60)

statistik_deskriptif = df[kolom_numerik].describe().T
print(statistik_deskriptif)

print("\nInterpretasi Kunci Angka Statistik:")
print("-" * 35)
print(f"1. **Survival Rate:** Rata-rata {statistik_deskriptif.loc['Survived', 'mean']*100:.2f}% selamat.")
print(f"2. **Usia (Age):** Rata-rata {statistik_deskriptif.loc['Age', 'mean']:.2f} tahun. Terdapat kemiringan ke kanan (skewness).")
print(f"3. **Tarif (Fare):** Terdapat **pencilan** tarif tinggi (Max ${statistik_deskriptif.loc['Fare', 'max']:.2f}).")


# ------------------------------------------------------------------------------
# 3. OBJEKTIF B: ANALISIS VISUAL UNTUK MENJAWAB PERTANYAAN
# ------------------------------------------------------------------------------

# Persiapan data untuk visualisasi tambahan
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['FamilyType'] = pd.cut(df['FamilySize'], bins=[0, 1, 4, 11], labels=['Sendiri', 'Kecil/Sedang', 'Besar'], right=True)

print("\n\n" + "="*60)
print("HASIL ANALISIS VISUAL UNTUK MENJAWAB PERTANYAAN KUNCI")
print("="*60)

fig, axes = plt.subplots(3, 2, figsize=(18, 15))
plt.suptitle('Analisis Visual Eksploratif Dataset Titanic', fontsize=20, y=1.02)


# Visualisasi 1: Survival Rate vs. Sex
sns.barplot(x='Sex', y='Survived', data=df, ax=axes[0, 0], palette='pastel')
axes[0, 0].set_title('1. Survival Rate Berdasarkan Jenis Kelamin', fontsize=14)

# Visualisasi 2: Survival Rate vs. Pclass
sns.barplot(x='Pclass', y='Survived', data=df, ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('2. Survival Rate Berdasarkan Kelas Penumpang', fontsize=14)

# Visualisasi 3: Distribusi/Sebaran Usia
sns.histplot(df['Age'].dropna(), kde=True, ax=axes[1, 0], bins=20, color='skyblue')
axes[1, 0].set_title('3. Distribusi Usia Penumpang', fontsize=14)

# Visualisasi 4: Survival Rate vs. Family Type (Menjawab Pertanyaan Baru)
sns.barplot(x='FamilyType', y='Survived', data=df, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('4. Survival Rate Berdasarkan Ukuran Keluarga', fontsize=14)

# Visualisasi 5: Korelasi Antar Kolom (Heatmap)
df['Sex_numeric'] = df['Sex'].apply(lambda x: 1 if x == 'female' else 0)
kolom_korelasi_heatmap = ['Survived', 'Pclass', 'Sex_numeric', 'Age', 'Fare', 'FamilySize']
korelasi_matrix = df[kolom_korelasi_heatmap].corr()

sns.heatmap(korelasi_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5, linecolor='black', ax=axes[2, 0])
axes[2, 0].set_title('5. Matriks Korelasi Variabel Kunci', fontsize=14)
axes[2, 0].set_xticklabels(axes[2, 0].get_xticklabels(), rotation=45, ha='right')
axes[2, 0].set_yticklabels(axes[2, 0].get_yticklabels(), rotation=0)

# Visualisasi 6: Distribusi Usia dan Survival berdasarkan Kelas
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=df, split=True, ax=axes[2, 1], palette={0: 'lightcoral', 1: 'lightgreen'})
axes[2, 1].set_title('6. Distribusi Usia dan Survival berdasarkan Kelas', fontsize=14)

plt.tight_layout()
plt.show()

# Bersihkan kolom temporer
df = df.drop(columns=['Sex_numeric', 'FamilySize', 'FamilyType'])


# ------------------------------------------------------------------------------
# 4. OBJEKTIF C: DOKUMENTASI INSIGHT DAN HIPOTESIS
# ------------------------------------------------------------------------------

print("\n\n" + "="*60)
print("DOKUMENTASI INSIGHT (TEMUAN) DAN HIPOTESIS AWAL")
print("="*60)
print("Berdasarkan analisis visual dan statistik, berikut adalah tiga temuan utama yang menjadi fondasi pemodelan:")

# --- INSIGHT 1 ---
print("\n" + "⭐ INSIGHT 1: PRIORITAS GENDER DAN KELAS")
print("Temuan Awal (Insight):")
print("Survival sangat kuat dipengaruhi oleh **Gender** (Wanita $\\approx 74\%$ selamat) dan **Pclass** (Kelas 1 $\\approx 64\%$ selamat). Kedua faktor ini menciptakan hierarki akses ke sekoci.")
print("Hipotesis Awal (H1):")
print("Fitur **'Sex'** dan **'Pclass'** akan menjadi prediktor paling dominan dalam model, menegaskan prioritas 'Women and Children First' dan keuntungan lokasi Kelas Atas.")

# --- INSIGHT 2 ---
print("\n" + "⭐ INSIGHT 2: EFEK UKURAN KELUARGA")
print("Temuan Awal (Insight):")
print("Visualisasi menunjukkan bahwa kelompok **Kecil/Sedang** (2-4 orang) memiliki peluang survival tertinggi, sementara penumpang **Sendiri** dan keluarga **Besar** lebih rentan.")
print("Hipotesis Awal (H2):")
print("Kita harus menggunakan fitur **kategori ukuran keluarga** (bukan jumlah anggota mentah) untuk menangkap efek **'sweet spot'** ini, di mana dukungan keluarga kecil meningkatkan survival, tetapi kompleksitas keluarga besar menghambatnya.")

# --- INSIGHT 3 ---
print("\n" + "⭐ INSIGHT 3: KERENTANAN USIA PER KELAS")
print("Temuan Awal (Insight):")
print("Di semua kelas, **anak-anak** memiliki tingkat survival yang lebih baik. Namun, di Kelas 3, konsentrasi besar penumpang **dewasa muda (20-30 tahun)** meninggal, menunjukkan bahwa faktor kepanikan/akses lebih mematikan bagi kelompok usia produktif di kelas bawah.")
print("Hipotesis Awal (H3):")
print("Pengolahan fitur **'Age'** harus dilakukan secara terpisah atau dalam interaksi dengan **'Pclass'** untuk menangkap perbedaan pola survival antara anak-anak di kelas atas dan dewasa di kelas bawah.")
```
